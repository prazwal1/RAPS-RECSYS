```python
import pandas as pd 
```

```python
df = pd.read_csv('olist_order_items_dataset.csv')
df[df.product_id == "4244733e06e7ecb4970a6e2683c13e61"]
```

```python
x = df.groupby('product_id')['order_item_id'].sum()
data = pd.DataFrame(x)
data.drop_duplicates(inplace = True)
data.sort_values(by = 'order_item_id',ascending = False)
```

```python
DF = pd.read_csv('final_features.csv')
DF
```

```python
y = pd.merge(DF,data,on='product_id',how='outer')
```

```python
y.fillna(0, inplace = True)
y.reset_index(inplace = True ,drop = True)
y.to_csv('final_features.csv',index = False)
```
