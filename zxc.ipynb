```python
vs = pd.read_csv('olist_products_dataset.csv')
sd = vs[['product_category_name']]
sd = sd.drop_duplicates()
y = []
for i in range (0,74):
    y.append(i)
yf = pd.DataFrame(y)
sd.reset_index(inplace = True, drop = True)
sd[['category_index']] = yf
vs['product_category_name'].fillna('others',inplace = True)
vs.fillna(0, inplace = True)
sd = sd.append({'product_category_name':'others', 'category_index' : 74}, ignore_index = True)
sd
```

```python
Database = pd.merge(sd,vs)
Database
```

```python
xsData = pd.merge(Database,data)
xsData = xsData[['product_id','category_index','seller_index']]
xsData 
```
