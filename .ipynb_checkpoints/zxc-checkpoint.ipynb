```python
import numpy
import pandas as pd 
```

```python
df = pd.read_csv('olist_order_items_dataset.csv')
df1 = pd.read_csv('olist_sellers_dataset.csv')
data = df[['product_id','seller_id']]
z = pd.merge(df1,data)
z.groupby('seller_id')
z
```

```python
x = []
for i in range (0,3095):
    x.append(i)
xf = pd.DataFrame(x)
df1[['seller_index']] = xf
df1
```

```python
data = pd.merge(df[['product_id','seller_id']],df1[['seller_id','seller_index']])
data
```

```python
vs = pd.read_csv('olist_products_dataset.csv')
sd = vs[['product_category_name']]
sd = sd.drop_duplicates()
y = []
for i in range (0,74):
    y.append(i)
yf = pd.DataFrame(y)
sd.reset_index(inplace = True, drop = True)
sd[['category_index']] = yf
sd
```

```python
Database = pd.merge(sd,vs)
Database
```

```python
xsData = pd.merge(Database,data)
xsData = xsData[['category_index','seller_index']]
vector = xsData.head(100).values
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
csim = cosine_similarity(vector)
csim
```
