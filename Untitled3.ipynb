```python
df = pd.read_csv('olist_sellers_dataset.csv')
df = df[['seller_id', 'seller_zip_code_prefix']]
df.rename(columns = {'seller_zip_code_prefix': 'geolocation_zip_code_prefix' }, inplace = True)
```

```python
data = pd.merge(df,df1, on = 'geolocation_zip_code_prefix')
```

```python
x = df[(~df.geolocation_zip_code_prefix.isin(data.geolocation_zip_code_prefix))&(~df.geolocation_zip_code_prefix.isin(data.geolocation_zip_code_prefix))]
```

```python
#data.drop_duplicates(inplace = True)
D1 = data.groupby('seller_id').max()
D1 
```

```python
def AddLatitude(row):
    code = row[['geolocation_zip_code_prefix']]
    code = int(code)
    if code == 82040:
        return 	10.777
    elif code == 91901:
        return 64.673
    elif code == 72580:
        return 19.0486
    elif code == 2285:
        return -32.907
    elif code == 7412:
        return 49.7014
    elif code == 71551:
        return 27.750198
    elif code == 37708:
        return 36.3249
    else:
        return 0
    
def AddLongitude(row):
    code = row[['geolocation_zip_code_prefix']]
    code = int(code)
    if code == 82040:
        return -74.8534
    elif code == 91901:
        return 25.6545
    elif code == 72580:
        return -98.4426
    elif code == 2285:
        return 151.6053
    elif code == 7412:
        return 6.0189
    elif code == 71551:
        return -97.420044
    elif code ==37708:
        return -83.3142
    else:
        return 0



```

```python
x['geolocation_lat'] = x.apply(lambda row : AddLatitude(row), axis = 1)
x['geolocation_lng'] = x.apply(lambda row : AddLongitude(row), axis = 1)
```

```python
x
```

```python
x['geo_label'] = x.apply(lambda row: AddIndex(row), axis=1)
x1 = x[['seller_id','geo_label']]
```

```python
x1.reset_index(inplace = True , drop = True)
x1
```

```python
def AddIndex(row):
    lattitude = row[["geolocation_lat"]]
    longitude = row[["geolocation_lng"]]
    lattitude = int(lattitude)
    longitude = int(longitude)

    if lattitude in range(-40,-30) and longitude in range(-80,-20):
        return 1
    elif lattitude in range(-30,-20) and longitude in range(-80,-20):
        return 2
    elif lattitude in range(-20,-10) and longitude in range(-10,-20):
        return 3
    elif lattitude in range(-10,0) and longitude in range(-80,-20):
        return 4
    elif lattitude in range(0,10) and longitude in range(-80,-20):
        return 5
    elif lattitude in range(20,-30) and longitude in range(-110,80):
        return 7
    elif lattitude in range(-20,30) and longitude in range(-20,0):
        return 8
    elif lattitude in range(-30,-20) and longitude in range(50,20):
        return 9
    else:
        return 12
```
