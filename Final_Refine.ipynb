```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('olist_products_dataset.csv')
df = df[['product_id','product_category_name','product_weight_g','product_length_cm','product_height_cm','product_width_cm']]
df.sort_values(by = 'product_id' , inplace = True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df1 = pd.read_csv('olist_order_items_dataset.csv')
df1 = df1.drop_duplicates()
df2 = df1[['product_id','price']]
df2 = df2.groupby('product_id',sort = False)['price'].mean()
data = pd.DataFrame(df2)
data = data.sort_values(by = 'product_id')
data
```

```python
Data = pd.merge(df,data.reset_index())
Data['product_category_name'].fillna('others',inplace = True)
Data.isna().any()
```
