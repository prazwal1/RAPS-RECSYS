```python
import pandas as pd
```

```python
df = pd.read_csv('olist_sellers_dataset.csv')
x = []
for i in range (0,3095):
    x.append(i)
df['seller_index'] = x
df = df[['seller_id' , 'seller_index']]
df
```

```python
df1 = pd.read_csv('olist_order_items_dataset.csv')
df1 = df1[['product_id','seller_id']]
df1.drop_duplicates(inplace = True)
df1.reset_index(drop = True)
data = pd.merge(df,df1)
data = data.groupby('product_id',sort = False)['seller_index'].max()
data
dataframe = pd.DataFrame(data)
dataframe
```

```python
Data = pd.read_csv('database_f.csv')
Data = Data.round(1)
Data
```

```python
xdata = pd.merge(Data,dataframe.reset_index())
xdata = xdata[['product_id','seller_index','product_category_name_english']]
xdata.isna().any()
```

```python
Df = pd.read_csv("ratings.csv")
Df.drop_duplicates()
```

```python
fdata = pd.merge(Df,xdata)
fdata
```

```python
categories = Data[['product_category_name_english']].drop_duplicates()
categories.reset_index(drop = True,inplace = True )
categories.reset_index(inplace = True)
categories.rename(columns = {"index" : "category_index"},inplace = True)
categories
```

```python
gdata = pd.merge(fdata.reset_index(),categories)

gdata = gdata[['product_id','review_score','seller_index','category_index']]
gdata
```

```python
gdata.to_csv('features.csv')
```
